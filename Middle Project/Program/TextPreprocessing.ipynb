# Load PubMed

```python
import pandas as pd
import numpy as np
import re
import nltk
import matplotlib.pyplot as plt
```

```python
PubMed_Articles_Location = "../Src/pubmed-CancerType_Top1-10-set_Output.xlsx"
PubMed_Articles = pd.read_excel(PubMed_Articles_Location)

PubMed_Articles.head()
```

```python
PubMed_Articles['Content'] = PubMed_Articles['Title']+ ' ' + PubMed_Articles['Abstract']
PubMed_Articles
```

```python
PubMed_Articles = PubMed_Articles.reindex(columns = ['PUMID','Title','Abstract','Content','CancerType'])
PubMed_Articles
```

```python
PubMed_Articles.info()
```

```python
ClassIDMap = {'Lung': 1,
              'Breast':2,
              'Colorectal':3,
              'Prostate': 4,
              'Stomach':5,
              'Liver':6,
              'Oesophagus':7,
              'Cervix Uteri':8,
              'Thyroid':9,
              'Bladder': 10}
ClassIDMap
```

```python
corpus, target_labels, target_names = (PubMed_Articles['Content'], [ClassIDMap[label] for label in PubMed_Articles['CancerType']], PubMed_Articles['CancerType'])
data_df = pd.DataFrame({'Article': corpus, 'Target Label': target_labels, 'Target Name': target_names})
print(data_df.shape)
data_df.head(10)
```

```python
filter_data = pd.DataFrame()
for i in data_df['Target Label'].unique():
    tmp = data_df[data_df['Target Label'] == i]
    test = tmp.sample(n=1000, random_state=1)
    filter_data = pd.concat([filter_data, test])
print(filter_data.head(5))
print(filter_data.shape)
```

# Text Preprocessing

```python
import text_normalizer as tn
import nltk
stopword_list = nltk.corpus.stopwords.words('english')
# just to keep negation if any in bi-grams
stopword_list.remove('no')
stopword_list.remove('not')
#stopword_list.add('cancer')

# normalize our corpus
norm_corpus = tn.normalize_corpus(corpus=filter_data['Article'], html_stripping=True, contraction_expansion=True, 
                                  accented_char_removal=True, text_lower_case=True, text_lemmatization=True, 
                                  text_stemming=False, special_char_removal=True, remove_digits=True,
                                  stopword_removal=True, stopwords=stopword_list)
filter_data['Clean Article'] = norm_corpus
```

```python
filter_data = filter_data[['Article', 'Clean Article', 'Target Label', 'Target Name']]
filter_data.head(10)
```

```python
filter_data = filter_data.replace(r'^(\s?)+$', np.nan, regex=True)
filter_data = filter_data.dropna().reset_index(drop=True)
```

```python
filter_data.to_csv('../Output/PubMed_CleanArticles_Top1-10_cancerTypes.csv', encoding='utf-8',index=False)
```
